```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
test_df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
train_df.head()
```

```python
train_df['target'] = train_df['target'].str.replace(',', '.')
```

```python
train_df.head()
```

```python
train_df['target'] = train_df['target'].astype(np.float64)
```

```python
test_df.head()
```

```python
test_df['target'] = test_df['target'].str.replace(',', '.')
```

```python
test_df.head()
```

```python
test_df['target'] = test_df['target'].astype(np.float64)
```

```python
test_df.head()
```

```python
train_df = train_df[train_df['target'] != 0.].copy()
test_df = test_df[test_df['target'] != 0.].copy()
```

```python
train_df.columns
```

```python
train_df['NDVI'] = (train_df['B08'] - train_df['B04']) / (train_df['B08'] + train_df['B04'])
test_df['NDVI'] = (test_df['B08'] - test_df['B04']) / (test_df['B08'] + test_df['B04'])
```

```python
y = train_df.target.values
X = train_df.drop(['target'], axis=1)

y_ = test_df.target.values
X_ = test_df.drop(['target'], axis=1)
```

```python
sc = StandardScaler()
sc.fit(X)
X = sc.transform(X)
X_ = sc.transform(X_)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_absolute_error
```

```python
rf = RandomForestRegressor(n_estimators=50, max_depth=4)
rf.fit(X, y)
p = rf.predict(X_)
print(f"Target min: {np.round(np.min(y_), 3)}\tTarget max: {np.round(np.max(y_), 3)}")
print(f"Mean Absolute Error: {np.round(mean_absolute_error(p, y_), 3)}")
print(f"Mean Absolute Percentage Error: {np.round(mean_absolute_percentage_error(p, y_), 3)}")
```
